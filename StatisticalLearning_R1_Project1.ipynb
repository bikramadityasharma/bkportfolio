Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. (2.5 points)

```python
import numpy as np
import pandas as pd
```

```python
la_df = pd.read_csv("Laliga.csv")
la_df.head()
```

```python
la_df.dtypes
```

```python
la_df.replace("-", 0, inplace = True)
la_df.describe()
```

```python
la_df = la_df.astype({'Points':'int64','GamesPlayed':'int64','GamesWon':'int64','GamesDrawn':'int64','GamesLost':'int64','GoalsFor':'int64','GoalsAgainst':'int64','Champion':'int64','Runner-up':'int64','Third':'int64','Fourth':'int64','Fifth':'int64','Sixth':'int64','T':'int64'})
#la_df.describe(include=np.object).transpose()
la_df.dtypes
la_df.describe().transpose()
```

```python
la_df["Debut"]= la_df["Debut"].str.split("-", n = 1, expand = True) 
print(la_df)
```

```python
la_df.dtypes
```

```python
la_df['Debut'] =  pd.to_datetime(la_df['Debut'],format='%Y')
```

```python
la_df.dtypes
```

Print all the teams which have started playing between 1930-1980. (5 points)

```python
mask = (la_df['Debut'] > '1930') & (la_df['Debut'] <= '1980')
print(la_df.loc[mask])
```

Print the list of teams which came Top 5 in terms of points (2.5 points)

```python
la_df['Points_Rank'] = la_df['Points'].rank(ascending = 0) 
la_df.head(5)
```

Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)
Goal_diff_count = GoalsFor - GoalsAgainst

```python
def goal_diff_count(goals_for,goals_against):
    goal_diff_count = goals_for -goals_against
    return goal_diff_count

la_df['GoalDiffCount'] = goal_diff_count(la_df['GoalsFor'],la_df['GoalsAgainst'])
print ("Minimum Goal Difference \n",la_df[la_df.GoalDiffCount==la_df.GoalDiffCount.min()])
print ("Maximum Goal Difference \n",la_df[la_df.GoalDiffCount==la_df.GoalDiffCount.max()])
```

Create a new column with name “Winning Percent” and append it to the data set (5 points)
Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

```python
la_df['WinningPercent'] = (la_df['GamesWon']/la_df['GamesPlayed'])*100
la_df['WinningPercent'].fillna(0, inplace=True)
print(la_df)
```
