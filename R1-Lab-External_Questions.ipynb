### Find out the no of male & female employees in the company

*Hint: Use can use a condition to find out the no of 1s and 0s in gender and sum it*

```python
gender = hr['Gender']
male=0
female=0
for i in gender:
    if i == 1:
        male=male+1
    elif i==2:
        female=female+1
print ("Assumption is that 1 stands for male and 2 for female in Gender column")        
print ("Male (1): ",male)
print ("Female (2): ",female)    

    
```

### Find out the IQR for the column - Age 

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
q1, q3= np.percentile(hr.Age, [25,75])
iqr = q3 - q1
print ("q1: {} , q3: {} , IQR: {}".format(q1,q3,iqr))
```
